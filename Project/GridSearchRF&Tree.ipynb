```python
import warnings
warnings.simplefilter('ignore')
import numpy as np      
import pandas as pd
credit = pd.read_csv("ccd.csv")
```

```python
from sklearn.model_selection import train_test_split
X,Y = credit.iloc[:,credit.columns != 'Approved'] , credit["Approved"]
credit.to_numpy
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=12)
```

```python
X_train.head()
```

<h1>Random Forest<h1>

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, Y_train)
y_pred = rf.predict(X_test)
print("Random Forest classifier has accuracy of: ", rf.score(X_test, Y_test))
```

<ul>
<li>Cross-validation</li>

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10, shuffle=True, random_state = 12)
ranFrost = RandomForestClassifier()
scores = cross_val_score(rf, X,Y,cv = kfold)
print("Cross-validation score for random Frost: \n{}".format(scores))
```

<li>Grid Search</li>

```python
from pprint import pprint
print("parameters currently in use for random forest:\n")
pprint(rf.get_params())
```

```python
n_estimators = [100, 200, 400, 800, 100]
max_features = ["auto", "sqrt"]
min_samples_split = [2,4,6,8,10]
min_samples_leaf = [1,2,3,4,5]
bootstrap = [True, False]
random_state = [12]

random_para_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap, 'random_state': random_state}
random_para_grid
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(ranFrost, random_para_grid, cv = 3, n_jobs = -1)
#x_trainval, X_val, y_trainval, y_val = train_test_split(X_train,Y_train, random_state=12)
rf_grid.fit(X_train,Y_train)
print(rf_grid.best_estimator_)
print(rf_grid.best_params_)
print(rf_grid.best_score_)
```

```python
print ("Test set score: {}".format(rf_grid.score(X_test, Y_test)))
```

<h2>Decision Tree<h2>

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree = tree.fit(X_train, Y_train)
print("Accuracy:", tree.score(X_test, Y_test))
```

<ul>
<li>Cross-validation</li>

```python
oneTree = DecisionTreeClassifier()
scores = cross_val_score(oneTree, X,Y, cv = 10)
print("Cross-validation score for desision tree: {}".format(scores))
```

<li>Grid Search</li>

```python
print("current use parameter in decision tree:\n")
pprint(tree.get_params())
```

```python
max_features = ["None", "auto"]
min_samples_split = [2,4,6,8,10]
min_samples_leaf = [1,2,3,4,5]

tree_para_grid = {
               'max_features': max_features,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
tree_para_grid
```

```python
tree_grid = GridSearchCV(oneTree, tree_para_grid, cv = 3, n_jobs = -1)
#x_trainval, X_val, y_trainval, y_val = train_test_split(X_train,Y_train, random_state=12)
tree_grid.fit(X_train,Y_train)
print(tree_grid.best_estimator_)
print(tree_grid.best_params_)
print(tree_grid.best_score_)
```

```python
print ("Test set score: {}".format(tree_grid.score(X_test, Y_test)))
```
