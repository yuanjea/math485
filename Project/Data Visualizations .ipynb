```python
x = credit['Age']
num_bins = 25
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(0,81,10))
plt.xlabel("age")
plt.ylabel("number of people")
plt.show
```

```python
x = credit['Approved']
num_bins = 2
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1)
plt.xticks(np.arange(0,1.1,1))
plt.xlabel("approved")
plt.ylabel("number of people")
```

```python
x = credit['Prior Default']
num_bins = 2
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1)
plt.xticks(np.arange(0,1.1,1))
plt.xlabel("PriorDefault")
plt.ylabel("number of people")
plt.show
```

```python
x = credit['Debt']
num_bins = 18
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-1,29,2))
plt.xlabel("Debt")
plt.ylabel("number of people")
```

```python
x = credit['Income']
num_bins = 10
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-10000,100000,5000),rotation=70)
plt.yticks(np.arange(0,701,50))
plt.xlabel("Debt")
plt.ylabel("number of people")
```

```python
x = credit['Credit Score']
num_bins = 18
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-5,66,5))
plt.yticks(np.arange(0,511,50))
plt.xlabel("Credit Score")
plt.ylabel("number of people")
```

```python
x = credit['Years Employed']
num_bins = 18
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-1,29,1),rotation=75)
plt.xlabel("year of employed")
plt.ylabel("number of people")
```
