# Data Analysis and Visualizations

```python
# Import Relevant Modules 
import warnings
warnings.simplefilter('ignore')
import numpy as np      
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
credit = pd.read_excel("credit_card_data.xlsx")
```

```python
# Dataset Statistics
credit.describe()
```

# Numerics

```python
# Number of Male = '1' in the dataset 
maleCount=0
for x in credit['Male']:
    if(x==1):
        maleCount+=1
maleCount
```

```python
# Number of Approved = '1' in the dataset
appCount=0
for x in credit['Approved']:
    if(x==1):
        appCount+=1
appCount
```

```python
# Number of Employed = '1' in the dataset
empCount=0
for x in credit['Employed']:
    if(x==1):
        empCount+=1
empCount
```

# Histogram Analysis
## Histogram of Age

```python
x = credit['Age']
num_bins = 25
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(0,81,10))
plt.xlabel("age")
plt.ylabel("number of people")
plt.show
```

## Histogram of Approved 

```python
x = credit['Approved']
num_bins = 2
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1)
plt.xticks(np.arange(0,1.1,1))
plt.xlabel("approved")
plt.ylabel("number of people")
```

## Histogram of Prior Default

```python
x = credit['Prior Default']
num_bins = 2
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1)
plt.xticks(np.arange(0,1.1,1))
plt.xlabel("PriorDefault")
plt.ylabel("number of people")
plt.show
```

## Histogram of Debt

```python
x = credit['Debt']
num_bins = 18
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-1,29,2))
plt.xlabel("Debt")
plt.ylabel("number of people")
```

## Histogram of Income

```python
x = credit['Income']
num_bins = 10
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-10000,100000,5000),rotation=70)
plt.yticks(np.arange(0,701,50))
plt.xlabel("Debt")
plt.ylabel("number of people")
```

## Histogram of Credit Score

```python
x = credit['Credit Score']
num_bins = 18
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-5,66,5))
plt.yticks(np.arange(0,511,50))
plt.xlabel("Credit Score")
plt.ylabel("number of people")
```

## Histogram of Years Employed

```python
x = credit['Years Employed']
num_bins = 18
n, bins, patches = plt.hist(x, num_bins, edgecolor='black', linewidth = 1, align = 'left')
plt.xticks(np.arange(-1,29,1),rotation=75)
plt.xlabel("year of employed")
plt.ylabel("number of people")
```

# Correlation Matrix Plot
Using Pearson correlation coefficient

Top 4 Correlated Attributes to the Approved attribute:
- Years Employed
- Prior Default
- Employed
- Credit Score

```python
credit_drop=credit.drop(["Zip Code"],axis=1)
corr = credit_drop.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, annot=True)
```
